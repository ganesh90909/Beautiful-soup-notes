```python
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get('https://www.imdb.com/list/ls009668766/')
soup=BeautifulSoup(page.content,'html.parser')
```

```python
page.status_code
```

```python
items=soup.find_all(class_='lister-item mode-detail')
movie1=items[0]
print(movie1.prettify())
```

```python
movie_title=movie1.h3.find('a').get_text()
print(movie_title)

movie_genre=movie1.find(class_='genre').get_text()
print(movie_genre)

imdb_rating=movie1.find(class_='ipl-rating-star__rating').get_text()
print(imdb_rating)

metascore=movie1.find(class_='metascore favorable').get_text()
print(metascore)

vote_class=movie1.find_all(class_='text-muted text-small')
vote_span=vote_class[2]
votes=vote_span.find_all('span')
print(votes[1].get_text())
```

## Collecting information of all the movies in the webpage

```python
Movies=[]
Genres=[]
IMDBratings=[]
Metascores=[]
totalVotes=[]



for movie in soup.find_all(class_='lister-item mode-detail'):
    movie_title=movie.h3.find('a').get_text()
    Movies.append(movie_title)
    
    movie_genre=movie.find(class_='genre').get_text()
    Genres.append(movie_genre)
    
    imdb_rating=movie.find(class_='ipl-rating-star__rating').get_text()
    IMDBratings.append(imdb_rating)
    
    if movie.find(class_='metascore'):
        metascore=movie.find(class_='metascore').get_text()
        Metascores.append(metascore)
    else:
        Metascores.append('0')
    
    vote_class=movie.find_all(class_='text-muted text-small')
    vote_span=vote_class[2]
    votes=vote_span.find_all('span')
    totalVotes.append(votes[1].get_text())
```
